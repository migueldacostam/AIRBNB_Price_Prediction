### 8. Valores únicos de todas las columnas de df_mine

```python
# Columnas específicas para obtener valores únicos
columns_of_interest = [
    'neighbourhood_group_cleansed',
    'property_type',
    'room_type',
    'accommodates',
    'bathrooms'
]

# Obtener y ordenar valores únicos de las columnas especificadas

for column in columns_of_interest:
    if column in df_mine.columns:
        unique_values = df_mine[column].unique()
        if pd.api.types.is_numeric_dtype(df_mine[column]):
            sorted_unique_values = sorted(unique_values, reverse=True)
        else:
            sorted_unique_values = sorted(unique_values)
        print(f"Valores únicos en la columna '{column}':")
        print(sorted_unique_values)
        print()
    else:
        print(f"La columna '{column}' no está en el DataFrame.")
```

# 3. Modificación de columnas

```python
df_mine.head()
```

### 1. neighbourhood_group_cleansed	

```python
# Cambio nombre por neighbourhood
df_mine.rename(columns={'neighbourhood_group_cleansed': 'neighbourhood'}, inplace=True)
```

```python
# Transformar datos (minúsculas, sin acentos, _....etc)
df_mine['neighbourhood'] = df_mine['neighbourhood'].str.lower().str.replace(' ', '_').str.replace('-', '_').str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('ascii')
```

### 2. property_type

```python
# Transformar datos (minúsculas, sin acentos, _....etc)
df_mine['property_type'] = df_mine['property_type'].str.lower().str.replace(' ', '_').str.replace('-', '_').str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('ascii')
```

### 3. room_type

```python
# Transformar datos (minúsculas, sin acentos, _....etc)
df_mine[['neighbourhood', 'room_type']] = df_mine[['neighbourhood', 'room_type']].apply(lambda col: col.str.lower().str.replace(' ', '_').str.replace('-', '_').str.replace('/', '_').str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('ascii'))
```

### 4. price

```python
# Transformar datos y pasar a columna numérica
df_mine['price'] = df_mine['price'].str.replace('[\$,]', '', regex=True).astype(float)
```
